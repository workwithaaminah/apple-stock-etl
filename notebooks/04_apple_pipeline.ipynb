```python
df_clean["Date"] = pd.to_datetime(df_clean["Date"])
```

```python
df_clean["Year"] = df_clean["Date"].dt.year
```

```python
df_clean.groupby("Year")["Date"].count()
```

```python
# step 1 group by year - Put rows from the same year together.
# step 2 - aggregate daily return pct by year using mean - Calculate the average daily return for each year.

yearly_returns = df_clean.groupby("Year")["daily_return_pct"].mean()

# step 3 - sort the results by daily return pct in descending order - Show the best years first.
# Sorting helps us identify rankings and priorities.
yearly_returns.sort_values(
    ascending=False
    ).head(10)

# Repeated workflow: Raw Data -> Group -> Aggregate -> Sort -> Report
```
