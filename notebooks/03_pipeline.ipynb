```python
import pandas as pd

df = pd.read_csv("../data/iris.csv")
```

```python
df_clean = df.copy()
```

```python
df_clean["sepal_area"] = (
    df_clean["sepal_length"]
    * df_clean["sepal_width"]
)
```

```python
df_clean.head()
```

```python
df_clean.to_csv(
    "../data/iris_clean.csv",
    index=False
)
```

```python
df_clean.isnull().sum()
```

```python
df_clean.duplicated().sum()
```

```python
print("Rows:", len(df_clean))
print("Missing values:")
print(df_clean.isnull().sum())

print("Duplicate rows:")
print(df_clean.duplicated().sum())
```

```python
df_clean["petal_area"] = df_clean["petal_length"] * df_clean["petal_width"]
```

```python
df_clean.head()
```

```python
df_clean.to_csv(
    "../data/iris_clean.csv",
    index=False
)
```

```python
df_clean["petal_to_sepal_ratio"] = ( 
    df_clean["petal_area"] / df_clean["sepal_area"]
    )
```
